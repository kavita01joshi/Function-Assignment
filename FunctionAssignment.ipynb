10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
*  `map()`, `reduce()`, and `filter()` are powerful higher-order functions in Python used for processing lists in a functional programming

 *- `map()` - Applies a function to each element
     * purpose : transforms all items in a list by applying a function.
     * output - A new list with modified values.
  
            l= [2, 7, 9, 21]
            squear = list(map(lambda x:x**2, l))
            print(squear) # applies the lambda function to each element in l

 *- `reduce()` - Combines all elements into a single value.
      * purpose : Applies a function repeatedly, reducing the list to one result.
      * output - A single value.

             from functools import reduce
             l= [2, 7, 9, 21]
            total = reduce(lambda x, y: x+y, l)
            print(total) # Output: 39


 *- `filter()` - Selects items that match a condition.
       * purpose : Removes unwanted elements based on a condition.
      * output - A new list with only the items that meet the condition.
      
            l= [2, 7, 9, 21]
            is_even = list(filter(lambda x: x%2==0, l))
            print(is_even) # Output: [2]

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.

class fib_iterator:
  def __init__(self, n):
    self.n = n
    self.a, self.b = 0, 1
    self.count = 0

  def __iter__(self):
    return self

  def __next__(self):
    if self.count >= self.n:
      raise StopIteration  # Stop when limit is reached
    value = self.a
    self.a, self.b = self.b, self.a + self.b  # Move forward in sequence
    self.count += 1
    return value

# Example usage:
for num in fib_iterator(5):  # First 5 Fibonacci numbers
    print(num)

```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def power_of_2(n):
  for i in range(n+1):
    yield 2**i
for power in power_of_2(7):
  print(power)
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
def read_file_generator(file_path):
  with open(file_path, 'r') as file:
    for line in file:
      yield line.strip()

for line in read_file_generator("impDoc.txt"):
    print(line)
```
